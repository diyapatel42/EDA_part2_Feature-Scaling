```python
'''
Feature Scaling: rescale the values present in the features. Converting the scale of different measurements, in to a single scale. Standardizes the whole dataset in one range.
In order for any columan to not be prioritized in ML given their data we need to use Feature scaling. Just a scaled down value while fature importance does not change.
Two main :  standardization, normalization
'''
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('kamyr-digester.csv')
df.info()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler

'''

Normalization: Scaling techniques in which values are shifted and scaled so they end up ranging between 0 and 1. It is also known and min-max scaling.
Standardization: Scaling techinique where values are centered aroudn the mean with a unit standard deviation. This means taht the mean of the attribute becomes 0 and the resultant distribution has a unit standard deviation.

If there are comuns that are irrelevant for your research just remove them.
'''

df.head()
```

```python
#show statistical info of all the numerical columns
df.describe().round(2)
```

```python
'''
Normalization
'''
#new data frame with just Y-Kappa, ChipRate
print(df.columns)
```

```python
df1 = pd.DataFrame(df, columns= ['Y-Kappa','ChipRate'])
df1.head(5)
```

```python
#impute any null values with mean values
df1['Y-Kappa'] = df1['Y-Kappa'].fillna(df1['Y-Kappa'].mean())
df1.isna().sum()
```

```python
#initiate function
scaler = MinMaxScaler()
normalized_df1 = scaler.fit_transform(df1)
print(normalized_df1)
```

```python
'''Standardization'''
#initiate function
scaler1 = StandardScaler()
normalized_df2 = scaler1.fit_transform(df1)
print(normalized_df2)
```
